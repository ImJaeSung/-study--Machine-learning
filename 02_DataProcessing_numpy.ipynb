- concatenate 함수는 리스트나 튜플을 input으로 받아 첫 번째 차원을 따라 배열을 연결함
- np.ones, np.zeros는 원하는 개수의 1과 0을 채운 배열을 만들어줌

```python
from sklearn.model_selection import train_test_split
```

```python
train_input, test_input, train_target, test_target = train_test_split(fish_data, fish_target) # fish_data를 앞에 두 변수에, fist_target을 뒤에 두 변수에 나누어 할당
                                                                                              # 샘플링 편향이 일어남
```

- train_test_split 함수는 전달되는 리스트나 배열을 비율에 맞게 훈련 세트와 테스트 세트로 알아서 섞어서 나누어 준다.
- stratify 매개변수에 target 데이터를 전달하면 클래스 비율에 맞게 데이터를 나누어준다.

```python
print(test_target)
```

```python
train_input, test_input, train_target, test_target = train_test_split(fish_data, fish_target, stratify = fish_target) 
# stratify 매개변수를 이용하여 클래스 비율(기본값은 25%)에 맞게 데이터를 나누어 샘플링 편향 해결
```

```python
print(test_target)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn = KNeighborsClassifier()
kn.fit(train_input, train_target)
kn.score(test_input, test_target)
```

```python
print(kn.predict([[25, 150]])) # 그림 상 도미쪽에 더 가깝게 있지만 빙어라고 예측
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 150, marker = '^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
distances, indexes = kn.kneighbors([[25, 150]]) # kneighbors method는 이웃까지의 거리와 이웃 샘플의 인덱스를 반환, 기본값은 5개
```

```python
display(distances)
display(indexes)
```

- kneighbors 메서드는 주어진 샘플에서 가장 가까운 이웃을 찾아 이웃까지의 거리와 이웃 샘플의 인덱스를 반환
- return_distance 매개변수를 bool형으로 지정하여 거리 

```python
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 150, marker = '^')
plt.scatter(train_input[indexes,0], train_input[indexes, 1], marker = 'D')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 축의 스케일을 동일하게 맞추기
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 150, marker = '^')
plt.scatter(train_input[indexes,0], train_input[indexes, 1], marker = 'D')
plt.xlim((0, 1000))
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 표준점수로 바꾸기
mean = np.mean(train_input, axis = 0)
std = np.std(train_input, axis = 0) # axis = 0 은 컬럼으로 계산
print(mean, std)
```

- z-score 는 각 특성값이 평균에서 표준편차의 몇 배만큼 떨어져 있는지는 나타냄
- 실제 특성값의 크기와 상관없이 동일한 조건으로 비교할 수 있음

```python
train_scaled = (train_input - mean) / std  # numpy broadcasting 기능
                                           # numpy 배열끼리만 가능
```

- 브로드캐스팅(broadcasting) 은 모든 행에 연산을 적용해주는 기능 
- numpy 배열 사이에 일어남

```python
# 수상한 도미 다시 표시하기
new = ([25, 150] - mean) / std   # 훈련 세트 기준으로 데스트 세트를 표준점수로 변환

plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(new[0], new[1], marker = '^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
kn.fit(train_scaled, train_target)

test_scaled = (test_input - mean)/std  # target 데이터는 스케일을 조정하는 대상이 아님 주의
kn.score(test_scaled, test_target)
```

- 테스트 세트도 훈련 세트의 평균과 표준편차로 변환해야 한다.
- 예측하는 새로운 데이터도 훈련 세트를 기준으로 변환해야 한다.
